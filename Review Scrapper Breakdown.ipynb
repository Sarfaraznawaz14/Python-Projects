# Step 2:

```python
# this variable returns the page source of the page we want to browse. 
# to view below script in the web page (rightclick anywhere on webpage --> go to page source)
flipkartPage = uClient.read()
```

```python
# using beautiful soap parse this data using below command. 
flipkart_html = bs(flipkartPage, "html.parser")
```

```python
flipkart_html
```

```python
# Now in above data find the class tag of the product you want to search by using findall from beautifulsoap and asign it to a variable
# "class": "_1AtVbE col-12-12" this tag can be found by hovering mouse on the product you want to target in website
```

```python
bigboxes = flipkart_html.findAll("div", {"class": "_1AtVbE col-12-12"})
```

```python
# bigboxes[4] is selected as this box helps us to select a perticular product link.
# 
bigboxes[4].div.div.div.a["href"]
```

```python
# this step will give us link of the product of which we want to scrap the reviews 

del bigboxes[0:3]
box = bigboxes[0]
productLink = "https://www.flipkart.com" + box.div.div.div.a['href']
```

```python
productLink
```

```python
# This variable holds the link of the product
prodRes = requests.get(productLink)
```

```python
# <Response [200]> this response means that the product is available
prodRes
```

```python
# here we are trying to reach the page and the location of reviews. beautifulsoap will parse through the page.
prod_html = bs(prodRes.text, "html.parser")
```

```python
prod_html
```

```python
print(prod_html)
```

```python
# now we find the class in which reviews are stored using findall, same as above
commentboxes = prod_html.find_all('div', {'class': "_16PBlm"})
```

```python
commentboxes
```

```python
commentboxes[0].div.div.div.p.text
```

```python
len(commentboxes)
```

```python
# this command gives us the name of the person who has given the review
# name, review, comment, stars can be easily located by hovering mouse on the webpage. (practice)
commentboxes[1].div.div.find_all('p', {'class': '_2sc7ZR _2V5EHH'}) # it is returing list
```

```python
# thus to extract text from it
commentboxes[1].div.div.find_all('p', {'class': '_2sc7ZR _2V5EHH'})[0].text
```

```python
# lets generalise the code

for i in range(len(commentboxes)-1):
    print(commentboxes[i].div.div.find_all('p', {'class': '_2sc7ZR _2V5EHH'})[0].text) #prints the name of person
```

```python
for i in range(len(commentboxes)-1):
    print(commentboxes[i].div.div.div.div.text) # prints the rating of the product
```

```python
for i in range(len(commentboxes)-1):
    print(commentboxes[i].div.div.div.p.text) # prints the review of the product
```

```python
for i in range(len(commentboxes)-1):
    #print(commentboxes[i].div.div.find_all('div', {'class': ''})) # prints the tag of the product
    comtag = commentboxes[i].div.div.find_all('div', {'class': ''})
    print(comtag[0].div.text)
    
```

```python
# lets put all this together

for i in range(len(commentboxes)-1):
    print(commentboxes[i].div.div.find_all('p', {'class': '_2sc7ZR _2V5EHH'})[0].text)  #prints the name of person
    print(commentboxes[i].div.div.div.div.text) # prints the rating of the product
    print(commentboxes[i].div.div.div.p.text) # prints the review of the product
    comtag = commentboxes[i].div.div.find_all('div', {'class': ''})
    print(comtag[0].div.text) # prints the tag of the product
```
